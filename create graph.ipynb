```python
import random
import pagerank
def create_graph(n_graphs,n_nodes,prob,rsp=0.15):
    graphs = {}
    graphs['rsp']=rsp
    graphs['prob'] = prob
    graphs['n_nodes'] = n_nodes
    graphs['n_graphs'] = n_graphs
    for G in range(n_graphs):
        d = {}
        for i in range(n_nodes):
            d[i] = {}
            for j in range(n_nodes):
                if i==j:
                    continue
                if random.random()<prob:
                    d[i][j]=1
        scores = pagerank.powerIteration(d, rsp=rsp, epsilon=0.0001, maxIterations=1000)
        graphs[G] = {}
        graphs[G]['x'] = d
        graphs[G]['y'] = scores.to_json()
    return graphs
```

```python
graphs = create_graph(2,500,0.5)
```

```python
import json
def save(filename,data):
    with open('input/'+filename+'.txt','w') as f:
        json.dump(data,f)
```

```python
def read_file(filename):
    with open('input/'+filename+'.txt','r') as f:
        data = json.load(f)
    return data
```

```python
save('prob='+str(5),graphs)
```
